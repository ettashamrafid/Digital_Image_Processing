```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('moon.tif')
plt.imshow(img)
```

```python
hist, bins= np.histogram(img.flatten(), 256, [0,256])
plt.plot(hist)
```

```python
graying = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
equ = cv2.equalizeHist(graying)
```

```python
hist, bins= np.histogram(equ.flatten(), 256, [0,256])
plt.plot(hist)
```

```python
ref_img =cv2.imread('ref.png')
plt.figure(figsize=(16, 12))
plt.imshow(ref_img)
```

```python
hist, bins= np.histogram(ref_img.flatten(), 256, [0,256])
plt.plot(hist)
```

```python
print(img.ndim, ref_img.ndim)
```

```python
from skimage.exposure import match_histograms
matched = match_histograms(img, ref_img,channel_axis=-1)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(12, 10),
                                    sharex=True, sharey=True)
for aa in (ax1, ax2, ax3):
    aa.set_axis_off()

ax1.imshow(img)
ax1.set_title('Source')
ax2.imshow(ref_img)
ax2.set_title('Reference')
ax3.imshow(matched)
ax3.set_title('Matched')
```
