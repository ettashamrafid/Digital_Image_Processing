```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img= cv2.imread('Qs3.jpg')
plt.imshow(img)
```

```python
img_blur= cv2.GaussianBlur(img, (3,3), 1.4)
plt.imshow(img_blur)
```

# (a)

```python
img_canny= cv2.Laplacian(img_blur, -1,ksize=5)
plt.imshow(img_canny, cmap='gray')
```

# (b)

```python
img_canny= cv2.Laplacian(img, -1,ksize=3)
plt.imshow(img_canny, cmap='gray')
```

# (c)

```python
img_canny_2= cv2.Canny(img, 100,200)
plt.imshow(img_canny_2, cmap='gray')
```
