```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread('disc.png')
img1=cv2.imread('disc.png')
imgs=np.concatenate((img,img1), axis=1)
cv2.imshow('Images', imgs)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Converting to GreyScale

```python
gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_img1=cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
```

```python
hist=cv2.calcHist(gray_img, [0], None, [256], [0,256])
hist1=cv2.calcHist(gray_img1, [0], None, [256], [0,256])
```

```python
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.title('Image1')
plt.xlabel('bins')
plt.ylabel('No. of Pixels')
plt.plot(hist)

plt.subplot(122)
plt.title('Image2')
plt.xlabel('bins')
plt.ylabel('No. of Pixels')
plt.plot(hist1)
```

## Equalizing the Histogram

```python
gray_img_eq= cv2.equalizeHist(gray_img)
gray_img1_eq= cv2.equalizeHist(gray_img1)

hist=cv2.calcHist(gray_img_eq, [0], None, [256], [0,256])
hist1=cv2.calcHist(gray_img1_eq, [0], None, [256], [0,256])
```

```python
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.title('Image1')
plt.xlabel('bins')
plt.ylabel('No. of Pixels')
plt.plot(hist)

plt.subplot(122)
plt.title('Image2')
plt.xlabel('bins')
plt.ylabel('No. of Pixels')
plt.plot(hist1)
```

```python
eq_img=np.concatenate((gray_img_eq, gray_img1_eq), axis=1)
cv2.imshow('Images', eq_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## CLAHE

```python
clahe=cv2.createCLAHE(clipLimit=40)
gray_img_clahe=clahe.apply(gray_img_eq)
gray_img1_clahe=clahe.apply(gray_img1)
comp_img=np.concatenate((gray_img_clahe, gray_img1_clahe), axis=1)
cv2.imshow('Images', comp_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Thresholding

```python
th=80
max_val=255

ret, o1= cv2.threshold(gray_img_clahe, th, max_val, cv2.THRESH_BINARY)
cv2.putText(o1, 'Thresh Binary', (10,30), cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),3, cv2.LINE_AA)

ret, o2= cv2.threshold(gray_img_clahe, th, max_val, cv2.THRESH_BINARY_INV)
cv2.putText(o2, 'Thresh Binary Inv', (10,30),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),3, cv2.LINE_AA)

ret, o3= cv2.threshold(gray_img_clahe, th, max_val, cv2.THRESH_TOZERO)
cv2.putText(o3, 'Thresh Tozero', (10,30), cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),3, cv2.LINE_AA)

ret, o4= cv2.threshold(gray_img_clahe, th, max_val, cv2.THRESH_TOZERO_INV)
cv2.putText(o4, 'Thresh Tozero Inv', (10,30), cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),3, cv2.LINE_AA)

ret, o5= cv2.threshold(gray_img_clahe, th, max_val, cv2.THRESH_TRUNC)
cv2.putText(o5, 'Thresh Trunc', (10,30), cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),3, cv2.LINE_AA)

ret, o6= cv2.threshold(gray_img_clahe, th, max_val, cv2.THRESH_OTSU)
cv2.putText(o6, 'Thresh OSTU', (10,30), cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),3, cv2.LINE_AA)

final=np.concatenate((o1, o2, o3), axis=1)
final1=np.concatenate((o4, o5, o6), axis=1)
cv2.imwrite('Image1.jpg', final)
cv2.imwrite('Image2.jpg', final1)
```

## Adaptive Thresholding

```python
thresh1= cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
thresh2= cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 31, 4)
thresh3= cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
thresh4= cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 4)

final=np.concatenate((thresh1, thresh2, thresh3, thresh4), axis=1)
cv2.imwrite('adaptive.jpg', final)
```

## OTSU Binarization

```python
ret, thresh1_otsu= cv2.threshold(gray_img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
cv2.imwrite('Otsu.jpg', thresh1_otsu)
```
