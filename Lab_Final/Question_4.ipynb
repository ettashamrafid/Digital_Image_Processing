```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage.exposure import match_histograms
```

```python
img= cv2.imread('Qs4.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
ref= cv2.imread('RefQs4.jpg')
plt.imshow(cv2.cvtColor(ref, cv2.COLOR_BGR2RGB))
```

# Histogram for the images

```python
img_hist= cv2.calcHist(img, [0], None, [256], [0,255])
plt.plot(img_hist)
```

```python
ref_hist= cv2.calcHist(ref, [0], None, [256], [0,255])
plt.plot(ref_hist)
```

# Histogram Matching

```python
matched= match_histograms(img, ref, channel_axis=-1)
plt.imshow(cv2.cvtColor(matched, cv2.COLOR_BGR2RGB))
```

# Showcasing Images

```python
output= np.concatenate([img, matched], axis=1)

plt.figure(figsize=(10,6))
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
```
