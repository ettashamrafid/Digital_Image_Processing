```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv2.imread('ref.png')
```

```python
neg_img= 255-img
```

```python
gamma=2.0
gamma_corrected= np.power(img/255.0, gamma)*255.0
gamma_corrected= np.uint8(gamma_corrected)
```

```python
log_transformed= (np.log1p(img)*30).astype(np.uint8)
```

```python
min_intensity=50
max_intensity=200
contrast_stretched= np.clip((img- min_intensity)*(255.0/(max_intensity- min_intensity)),0,255).astype(np.uint8)
```

```python
titles=['Original','Negative', 'Gamma Correction', 'Logarthmic', 'Contrast Stretched']
images=[img, neg_img, gamma_corrected, log_transformed, contrast_stretched]
```

```python
for i in range(len(titles)):
    plt.subplot(2,3,i+1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis('off')
```
