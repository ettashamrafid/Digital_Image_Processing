### What are the differences between a linear and non-linear filter? In which cases does one perform better than the other?

Linear filtering is the filtering method where the value of output pixel is linear combinations of the neighboring input pixels. It can be done with convolution operation. For example, mean/average filters or Gaussian filter.
Non-linear filter is a filter whose output is not a linear function of its input. Non-linear filtering cannot be done with convolution or Fourier multiplication. Median filter is a simple example of a non-linear filter.

Linear filtering is used for smoothing out or preserving edge whereas non linear filtering is better at edge detection.

### Find the attached image, 2.png. The left one is the input image, and the right one is the output image. What kind of noise is induced in the input image? How can we achieve the output image? 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter, ImageEnhance
```

```python
img=cv2.imread('2.png')
img=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
kernel_size= 3
new_img= cv2.medianBlur(img, kernel_size)
```

```python
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB))

plt.subplot(122)
plt.imshow(cv2.cvtColor(new_img, cv2.COLOR_HSV2RGB))
```

### To apply a filter, we need to concentrate on two criteria. i. edge preservation; ii. noise reduction (denoising). Propose a filter or name two existing filters that can perform both in image processing. Write your code and justify your answer.

```python
img= cv2.imread('2.png')
bilateral= cv2.bilateralFilter(img, 15,75,75)
gaussian=cv2.GaussianBlur(img, (5,5), 0)
median=cv2.medianBlur(img, 3)
```

```python
plt.subplot(221)
plt.imshow(img)

plt.subplot(222)
plt.imshow(bilateral)

plt.subplot(223)
plt.imshow(gaussian)

plt.subplot(224)
plt.imshow(median)
```

### Find the attached image, 4.png. How can we achieve this kind of output in image processing? Take two input images and show the output as given in the image.

```python
from pymatting import *
```

```python
scale = 1.0

image = load_image("lemur.png", "RGB", scale, "box")
trimap = load_image("tripmap.png", "GRAY", scale, "nearest")

alpha = estimate_alpha_cf(image, trimap)
new_background = load_image("beach.png", "RGB", scale, "box")

foreground, background = estimate_foreground_ml(image, alpha, return_background=True)
new_image = blend(foreground, new_background, alpha)

images = [image, trimap, alpha, new_image]
grid = make_grid(images)
save_image("new.png", grid)
```
