```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img= cv2.imread('Qs2.png')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img_grey=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_grey, cmap='gray')
```

```python
img_blur= cv2.GaussianBlur(img_grey, (3,3), 1.4)
plt.imshow(cv2.cvtColor(img_blur, cv2.COLOR_BGR2RGB))
```

# (a)

```python
img_canny=cv2.Canny(img_blur, 50,100)
plt.imshow(img_canny, cmap="gray")
```

# (b)

```python
img_laplacian=cv2.Laplacian(img_blur, -1, ksize=3)
plt.imshow(img_laplacian, cmap="gray")
```

# (c)

```python
img_sobel_xy= cv2.Sobel(img_blur, -1,1,2,ksize=5)
plt.imshow(img_sobel_xy, cmap='gray')
```
